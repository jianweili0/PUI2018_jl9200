# IDEA:
#### young women  are more likely than young men to choose biking in weekends 



## NULL HYPOTHESIS:

#### young women  are less likely than young men to bike on weekends 

The fraction of young women ( age under 35) riding on weekend of all young women biking on any day of the week are less or equal to the proportion of young men (age under 35) on the weekends of all young men biking on any day of the week.


** Based on my observation of in NYC during the weekend, I came across lot of young women on the weekend, while during the weekday, I saw mostly young men commuting using citibike.


$H_0$ : $\frac{W_{\mathrm{weekend35}}}{W_{\mathrm{anyday35}}} <= \frac{M_{\mathrm{weekend35}}}{M_{\mathrm{anyday35}}} $

$H_a$ : $\frac{W_{\mathrm{weekend35}}}{W_{\mathrm{anyday35}}} > \frac{M_{\mathrm{weekend35}}}{M_{\mathrm{anyday35}}} $


I will use a significance level  $\alpha=0.05$

## Figure 1a: Distribution of Citibike bikers by gender in Feburary 2015, absolute counts



There are errors of the course? so? what is the error on the count statistics?

for N counts the error is sqrt(N) - this is because counts follow the Poisson statistic

1. I notices that in the gender there are some values of 0


```python
fig = pl.figure(figsize = (15,15))

counts_w = df['date'][df.gender ==2].groupby(df['daynum']).count()
norm_w = 1
error_w = np.sqrt(counts_w)
ax = ((counts_w)/norm_w).plot(kind="bar",color='Orange',  yerr=[
            ((error_w) / norm_w, (error_w) / norm_w)], label='women bikers')

counts_m = df['date'][df.gender==1].groupby(df['daynum']).count()
norm_m = 1
error_m = np.sqrt(counts_m)
ax = ((counts_m)/norm_m).plot(kind="bar",color='SteelBlue', alpha = 0.5, yerr=[
            ((error_m) / norm_m, (error_m) / norm_m)], label='men bikers')


xticklabels =['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
tmp = ax.set_xticklabels(xticklabels,rotation = 30,fontsize = 20)

ax.set_xlabel("Day of the week",fontsize = 15)
ax.set_ylabel("Ridership",fontsize = 15)
pl.legend(['Women bikers','Men bikers'],fontsize = 15)
```
