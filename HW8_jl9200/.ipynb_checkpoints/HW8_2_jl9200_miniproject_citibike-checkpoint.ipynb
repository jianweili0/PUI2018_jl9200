# IDEA:
#### young women  are more likely than young men to choose biking in weekends 



## NULL HYPOTHESIS:

#### young women  are less likely than young men to bike on weekends 

The fraction of young women ( age under 35) riding on weekend of all young women biking on any day of the week are less or equal to the proportion of young men (age under 35) on the weekends of all young men biking on any day of the week.


** Based on my observation of in NYC during the weekend, I came across lot of young women on the weekend, while during the weekday, I saw mostly young men commuting using citibike.


$H_0$ : $\frac{W_{\mathrm{weekend35}}}{W_{\mathrm{anyday35}}} <= \frac{M_{\mathrm{weekend35}}}{M_{\mathrm{anyday35}}} $

$H_a$ : $\frac{W_{\mathrm{weekend35}}}{W_{\mathrm{anyday35}}} > \frac{M_{\mathrm{weekend35}}}{M_{\mathrm{anyday35}}} $


I will use a significance level  $\alpha=0.05$

### Z Test

$ H_0 = P_0 - P_1 <=0 $

$P_0  = \frac{W_{\mathrm{weekend35}}}{W_{\mathrm{anyday35}}} $

$P_1 = \frac{M_{\mathrm{weekend35}}}{M_{\mathrm{anyday35}}} $


## The statistics is: 
# $z = \frac{(p_0 - p_1)}{SE} $
# $p =\frac{p_0  n_0 + p_1  n_1}{n_0+n_1}$
# $SE = \sqrt{ p  ( 1 - p )  (\frac{1}{n_0} + \frac{1}{n_1}) }$
## and in this notation 
# $p_0 = \frac{W_{\mathrm{weekend}}}{W_{\mathrm{total}}}$
### and 
# $p_1 = \frac{M_{\mathrm{weekend}}}{M_{\mathrm{total}}}$

```python
alpha = 0.05
p0 = sum(counts_w[-2:])*1.0/dataNorm_w

p1 = sum(counts_m[-2:])*1.0/dataNorm_m

n0 = dataNorm_w
n1 = dataNorm_m

print(p0,p1,n0,n1) 
```

```python
p = lambda p0, p1, n0,n1:(p0*n0 + p1*p1)/(n0+n1)

se = lambda p, n0, n1: np.sqrt(p*(1-p)*(1.0/ n0 + 1.0/ n1))


zscore  = lambda p0, p1, se: (p0-p1)/se

z = zscore(p0,p1,se(p(p0,p1,n0,n1),n0,n1))
print(z)
```

```python
# we see our z score is too high
p_2y = 1 - 0.9984

def report_result(p,a):
    print ('is the p value ' + 
           '{0:.2f} smaller than the critical value {1:.2f}?'.format(p,a))
    if p < a:
        print ("YES!")
    else: 
        print ("NO!")
    
    print ('the Null hypothesis is {}'.format(\
                            'rejected' if p < a  else 'not rejected') )
    
report_result(p_2y,alpha)
```

# Chi-Square

```python
# learning from Federica's Lab5 evalChisq.py

def evalChisq(values):
    '''
    Evaluates the chisq from a contigency value
    Arguments:
    values: 2X2 array or list, the contingengy table
    '''
    
    if not(len(values.shape) ==2 and values.shape ==(2,2)):
        print ("please enter 2D array")
        return -1
    values = np.array(values)
    E = np.empty_like(values)
    for j in range(len(values[0])):
        for i in range(2):
            
            E[i][j]=((values[i,:].sum()*values[:,j].sum())/
                    (values).sum())
        #print (E)
    return ((values -E)**2/ E).sum()
```

| **Rides on Weedend**	|  **Yes**  |  **No** | ** Sum ** |
|:----------:|:----------:|:------------:|:------------:|
|**Young Women** |p0 * n0 | p0_1 * n0| n0 |
|**Young Women** |p1 * n1 | p1_1 * n1| n1 |
|** Total ** | p0* n0 + p1 * n1 |  p0_1 * n0 + p1_1 * n1 | ntot |


```python
# total observations
ntot = n0+n1
# numbe of freedam = number of observations - 1
dof  = 2 -1
p0_1 = 1-p0
p1_1 = 1-p1
value = np.array([[p0*n0,p0_1*n0],[p1*n1,p1_1*n1]])
```

```python
cq = evalChisq(value)
print('Chi-Square score is {0:.3f}'.format(cq))
```
