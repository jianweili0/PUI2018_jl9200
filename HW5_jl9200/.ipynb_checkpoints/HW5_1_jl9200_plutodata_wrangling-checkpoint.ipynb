## Use groupby to count the building per LandUse code

```python
# reference https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
# df.groupby('column1',as_index = False)['column2'].count()


mngrouped = mn_data.groupby('LandUse',as_index = False)['ZipCode'].count()
bkgrouped = bk_data.groupby('LandUse',as_index = False)['ZipCode'].count()
```
