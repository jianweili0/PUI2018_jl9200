```python
# tradeoff of readibility and efficience 
ReprRandAll = np.zeros((50,2,100))
for i in range(50):
    ReprRandAll[i]=((np.random.randn(2,100)*np.random.randn(2,1)*5)+np.random.randn(2,1)*50)
    # let's unpack this: 
    # ReprRandAll[i] = ((np.random.randn(2, 100))) creates one 2x100 
    # Gaussian distributions with mean 0 and standard deviation 1 
    # (by default, see numpy documentation)
    #
    # ReprRandAll[i] = ((np.random.randn(2, 100) * np.random.rand(2, 1) * 5))
    # creates 2x100 Gaussian distributions then increases the standard deviation 
    # (the stretch of the distribution) by a random number between 0 and 5 
    # in each direction (x and y, different numbers in each direction). 
    # Now you have a 2D distribution of 100 points 
    # where each dimension is Gaussian distributed w mean 0 and 
    # stdev between 0 and 5
    # 
    # Finally + np.random.rand(2, 1) * 50 adds a random number between 0 and 50 
    # to each dimension, thus shifting the mean of the distribution along each axis
    # by a different amount (some amount between 0 and 50)
    #
    # note that randn generates Gaussian while rand Generated uniform distrubitions


ReprRandAll.shape
```

```python
###
import pylab as pl
%pylab inline
# the last line of code makes the plot show up within the notebook. 
# Always include it so that your homework notebooks are rendered and we can see
# what you wanted your results to be. 

# code here
pl.figure(figsize=(8,8))
for i in range(50):
    pl.plot(ReprRandAll[i,0], ReprRandAll[i,1], '.', alpha=0.3)
    pl.xlabel("x value", fontsize=20)
    pl.ylabel("y value", fontsize=20)
    pl.axes().set_aspect("equal")
```
